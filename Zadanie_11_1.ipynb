```python
import pandas as pd
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
import plotly.offline as pyo
import plotly.graph_objs as go
from plotly import subplots
```

```python
kghm = pd.read_csv('kgh_d.csv', index_col=0, parse_dates=True)
kghm.head()
```

```python
cu = pd.read_csv('ca_c_f_d.csv', index_col=0, parse_dates=True)
cu.head()
```

```python
fig = subplots.make_subplots(rows=2,cols=1,subplot_titles=['KGHM','Miedź'],shared_xaxes=True)

fig.add_trace(go.Scatter(x=kghm.index , y=kghm['Zamkniecie']), 1, 1)

fig.add_trace(go.Scatter(x=cu.index , y=cu['Zamkniecie']), 2, 1)

fig.update_layout(height=500, showlegend=False)

iplot(fig)
```

```python
fig = go.Figure(data=[go.Table(
    header=dict(values=['Data', 'KGHM', 'Miedź']),
    cells=dict(values=[kghm.index.strftime('%Y-%m-%d'), kghm['Zamkniecie'], cu['Zamkniecie']]))
])

fig.show()
```
